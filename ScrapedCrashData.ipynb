```python
import pandas as pd
url = 'https://en.wikipedia.org/wiki/List_of_aircraft_accidents_and_incidents_resulting_in_at_least_50_fatalities' 
tables = pd.read_html(url)
df = tables[3]
df.head(20) 
```

```python
structured_df=df.iloc[2:550,:]
structured_df=structured_df.rename(columns={0:'Total Deaths',
                         1:'Crew Deaths',
                         2:'Passenger Deaths',
                         3:'Ground Deaths',
                         4:'Notes',
                         5:'Aircraft Type',
                         6:'Incident / Airline',
                         7:'Aircraft',
                         8:'Location',
                         9:'Flight Phase',
                         10:'Airport',
                         11:'Distance',
                         12:'Date'})
structured_df.head(10)
```

```python
structured_df['Carrier'] = (structured_df['Aircraft']).str.split(n=1,expand=True)
structured_df.head()
```
